```python
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dict_view2 = {1: 1, 2: 1.33, 3: 1.74, 4: 1.96, 5: 2.17, 6: 2.37, 7: 2.70, 8: 2.79}
dict_view1 = {1: 1, 2: 1.24, 3: 1.62, 4: 1.58, 5: 2.09, 6: 2.23, 7: 2.43, 8: 3.01}
```

```python
sns.lineplot(dict_view1, label="View1")
sns.lineplot(dict_view2, label="View2")

plt.title("Speedup with multi-threads")
plt.xlabel("Number of threads")
plt.ylabel("Speedup")
plt.legend()
```
